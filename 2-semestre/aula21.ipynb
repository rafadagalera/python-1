```python
falhas = pd.read_csv("manutencao_preditiva.csv")
falhas
```

```python
# Crie um programa para somar as variaveis temperatura do ar e processo
# e retorne uma nova variavel
falhas["Variavel nova"] = falhas.apply(lambda row: row["Temperatura Ar [K]"] + row["Temperatura Processo [K]"],axis = 1)
falhas["Variavel nova"]
```

```python
# Faça a soma das temperaturas do ar e do processo apenas para as máquinas do tipo M
df_tipo_m = falhas[falhas["Tipo"] == "M"]
df_tipo_m["Soma temperatura"] = df_tipo_m.apply(lambda linha: linha["Temperatura Ar [K]"]+ linha["Temperatura Processo [K]"],axis = 1)
df_tipo_m["Soma temperatura"]
```

```python
df_tipo_m
```

```python
df_tipo_m["Soma temperatura"]
```

```python
# Mostre qual a média, mediana, min e max das temperaturas quando houver apenas
# maquinas do tipo M

df_tipo_m["Media"] = df_tipo_m["Soma temperatura"].apply(lambda x: x).mean()
df_tipo_m["Mediana"] = df_tipo_m["Soma temperatura"].apply(lambda x: x).median()
df_tipo_m["Max"] = df_tipo_m["Soma temperatura"].apply(lambda x: x).max()
df_tipo_m["Min"] = df_tipo_m["Soma temperatura"].apply(lambda x: x).min()
df_tipo_m
```

```python
# aplicar apply para verificar comprimentos de dados (strings)

df = pd.DataFrame({
    "Texto":["Hawaian","Roller","Coaster","Ride"]
})
df["contagem"] = df["Texto"].apply(lambda texto: len(texto))
df["contagem"]
```

```python
# Cria uma condição para inserir uma coluna chamada comprimento_texto e se o
# comprimento for maior que 5, identifique como texto longo, caso contrario
# identifique como texto curto

df["comprimento_texto"] = df["Texto"].apply(lambda texto: "Longo" if len(texto) >= 5 else "Curto" )
df
```

```python
# Crie um programa para verificar o comprimento do texto na coluna Tipo de Falha
# Após isso,insira uma nova coluna mostrando. Se o comprimento for menor ou igual
# a 10 - curto. Se for maior que 10 e menor que 15 - médio. Se for maior que 15 - longo


dados = pd.read_csv("manutencao_preditiva.csv")
dados["Verificando comprimento"] = dados["Tipo da Falha"].apply(lambda texto: "Curto" if len(texto) <= 10 else "Médio" if len(texto) > 10 and len(texto) >= 15 else "Longo")
dados["Verificando comprimento"].value_counts()

```

```python
# utilizando groupby do pandas

df = pd.DataFrame({
    "var_1":["A","A","B","A","C","C","A","B","A","B",],
    "var_2":[100,122,156,189,50,63,9,15,78,10],
    "var_3":[10,22,16,89,5,3,9,15,8,10]
})
agrupado_media = df.groupby("var_1").mean().reset_index()
agrupado_mediana = df.groupby("var_1").median().reset_index()
print(agrupado_media)
print(agrupado_mediana)
```

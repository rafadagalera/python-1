```python
# Faça a iteração usando iterrows para o dataframe falhas de máquinas
data = pd.read_csv("manutencao_preditiva.csv")
```

```python
# 1 Print as variáveis UDI, ID Produto e tipo 
for i, r in data.iterrows():
    print(f"Linha {i}")
    print(f"UDI: ",{r["UDI"]})
    print(f"ID Produto",{r["ID Produto"]})
    print(f"Tipo: ",{r['Tipo']})
```

```python
# 2 Como utilizar estrutura de condição usando iterrows?
for i,r in data.iterrows():
    if r["Tipo"] == "M":
        print(f"Linha {i}")
        print(f"UDI: ",{r["UDI"]})
        print(f"ID Produto",{r["ID Produto"]})
        print(f"Tipo: ",{r['Tipo']})
```

```python
# 3 Selecione apenas os dados do tipo H e print variáveis UDI, ID produto
#  e tipo
for i,r in data.iterrows():
    if r["Tipo"] == "H":
        print(f"Linha {i}")
        print(f"UDI: ",{r["UDI"]})
        print(f"ID Produto",{r["ID Produto"]})
        print(f"Tipo: ",{r['Tipo']})
```

```python
# 4
#  Selecione apenas os dados do tipo H e M e print as variáveis
#  Temperatura do Ar, processo, velocidade de rotação e torque

for i,r in data.iterrows():
    if r["Tipo"] == "H" or r["Tipo"] == "M":
        print(f"Linha {i}")
        print(f"Temperatura do Ar: ",{r['Temperatura Ar [K]']})
        print(f"Temperatura do Processo: ",{r['Temperatura Processo [K]']})
        print(f"Velocidade de Rotação: ",{r['Velocidade Rotacao [rpm]']})
        print(f"Torque: ", {r['Torque [Nm]']})
```

```python
# 5
# Crie um programa para mostrar um texto este é um tipo de máquina M 
# e se a máquina for do tipo h escreva essa máquina é do tipo h 
# caso contrário escreva essa máquina é de outro tipo

for i, r in data.iterrows():
    if r["Tipo"] == "M":
        print(f"Essa máquina é do tipo: ",{r["Tipo"]})
    elif r["Tipo"] == "H":
        print(f"Essa máquina é do tipo: ",{r["Tipo"]}) 
    else:
        print("Essa máquina tem um outro tipo!")    
```

```python
# 6
# Mostre as mesmas informações do exercício anterior, porém eslecione apenas 
# temperatura do ar maiores que 300 
for i, r in data.iterrows():
    if r['Temperatura Ar [K]'] > 300:
            if r["Tipo"] == "M":
                print(f"Essa máquina é do tipo: ",{r["Tipo"]})
            elif r["Tipo"] == "H":
                print(f"Essa máquina é do tipo: ",{r["Tipo"]}) 
            else:
                print("Essa máquina tem um outro tipo!")  
```

```python
# 8
# Crie uma estrutura para mostrar as informações de velocidade de rotação e torque.
# Porém, escreva temperatura alta de processo quando a temperatura for maior 
# do que a média e temperatura baixa quando for menor do que a média

media = data["Temperatura Processo [K]"].mean()

for i, r in data.iterrows():
    print(f"Velocidade de Rotação: ",{r['Velocidade Rotacao [rpm]']})
    print(f"Torque: ",{r['Torque [Nm]']})
    if r['Temperatura Processo [K]'] > media:
        print("Temperatura maior que a média")
    else:
        print("Temperatura menor que a média")
```

```python
# 9
# Crie um programa para indentificar e mostrar a quantidade de falhas
# no conjunto de dados

contagem_falhas = {}

for i, r in data.iterrows():
    tipo_falha = r["Tipo da Falha"]

    if tipo_falha in contagem_falhas:
        contagem_falhas[tipo_falha] += 1
    else:
        contagem_falhas[tipo_falha] = 1  

for tipo_falha, contagem in contagem_falhas.items():
    print("Tipo da falha: ", tipo_falha, "\n Quantidade: ", contagem)
```

```python
data = pd.read_csv("manutencao_preditiva.csv")
```

```python
data.columns
```

```python
# 11
# crie um programa para apresentar as informações de UDI e ID Produto
# quando houver apenas falhas do tipo no failure
iterator = data.iterrows()

while True:
    try:
        i, r = next(iterator)
        if r["Tipo da Falha"] == "No Failure":
            print(f"UDI: {r['UDI']}\n ID Produto: {r['ID Produto']} ")
        else:
            pass        
    except StopIteration:
        break    
```

```python
# 12
# Crie um programa para apresentar a quantidade de falhas no conjunto de dados
# de falhas em máquinas usando a estrutura while

contagem_falhas = {}
iterator = data.iterrows()
while True:
    try:
        i,r = next(iterator)
        tipo_falha = r["Tipo da Falha"]

        if tipo_falha in contagem_falhas:
            contagem_falhas[tipo_falha] += 1
        else:
            contagem_falhas[tipo_falha] = 1        
    except StopIteration:
        break
    
for tipo_falha, contagem in contagem_falhas.items():
    print("Tipo da falha: ", tipo_falha, "\n Quantidade: ", contagem)
```

```python
# 13
# Calcule a média usando for e while 

soma_desgaste = 0
count_m = 0

for i, r in data.iterrows():
    if r["Tipo"] == "M":
        soma_desgaste += r['Desgaste Ferramenta [min]']
        count_m += 1

if count_m > 0:
    media_desgaste = soma_desgaste / count_m
    print("A média de desgaste da ferramenta para o tipo M é ",media_desgaste)        
else:
    print("Não há dados")    
```

```python
soma_desgaste = 0
count_m = 0
iterator = data.iterrows()
while True:
    try:
        i,r = next(iterator)
        if r["Tipo"] == "M":
            soma_desgaste += r['Desgaste Ferramenta [min]']
            count_m += 1    
    except StopIteration:
        break

if count_m > 0:
    media_desgaste = soma_desgaste / count_m
    print("A média de desgaste da ferramenta para o tipo M é ",media_desgaste)        
else:
    print("Não há dados")    
```

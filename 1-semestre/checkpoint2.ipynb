```python








# Exercício 1: Para as três listas: itens, estoques e preços, utilize for para trazer os produtos cuja relação preço * estoque esteja acima de 100.

# Declarando as listas fornecidas.
itens = ['feijão','leite','açúcar','arroz','azeitona','café em grãos']
estoque = [50,20,30,50,30,90]
preços = [5.0,2.0,6.0,10.0,3.0,3.0]

# Multiplicando o número de itens em estoque pelo seu preço e os armazenando em uma nova lista "combine".
combine = [ b * c for b,c in zip(estoque,preços)]
# Criando uma lista vazia que armazenará os índices dos produtos que satisfazem a condição.
result = []
# Criando um loop que percorre a lista de relações preço * estoque, e adiciona o índice dos produtos que satisfazem a condição na lista result.
for i in range(len(combine)):
    if combine[i] > 100:
        result.append(i)
    else:
        pass
# Criando um loop que imprime os produtos cujos respectivos índices satisfazem as condições.   
for i in range(len(itens)):
    if i in result:
        print('O produto',itens[i], 'possui uma relação preço * estoque maior que 100.')
    else:
        pass    

        
   
    
```

```python
#Exercício 3: Você é um analista de dados de uma empresa e recebeu quatro listas de contendo informações sobre as vendas dos últimos meses.
#a) Calcule o total de vendas de cada produto ao longo dos quatro meses.
#b) Determine o produto mais vendido em cada mês
#c) Identifique o produto que teve o maior crescimento percentual nas vendas de janeiro para fevereiro.
#d) Liste os produtos que tiveram uma queda nas vendas de março para abril.

# Declarando as listas fornecidas.
produtos = ['A','B','C','D','E','F','G']
vendas_janeiro = [150,200,100,300,250,230,250]
vendas_fevereiro = [120,180,150,280,200,210,240]
vendas_março = [200,220,170,320,250,230,315]
vendas_abril = [180,215,185,290,250,410,390]
# a)
# Criando 4 novas variáveis que correspondem à soma dos valores das respectivas listas.
total_vendas_janeiro = sum(vendas_janeiro)
total_vendas_fevereiro = sum(vendas_fevereiro)
total_vendas_março = sum(vendas_março)
total_vendas_abril = sum(vendas_abril)
# Imprimindo o resultado em 4 linhas diferentes.
print('Total de vendas Janeiro:',total_vendas_janeiro,'\nTotal de vendas Fevereiro:', total_vendas_fevereiro, '\nTotal de vendas Março:', total_vendas_março,'\nTotal de vendas Abril:', total_vendas_abril)
#b)
# Criando uma lista vazia onde serão adicionados os índices das maiores vendas.
maiores_vendas = []
# Criando 4 loops, um para cada mês que funcionam de maneira idêntica. O loop percorre as listas de vendas respectivas de cada mês e adiciona o
# índice do maior valor da lista à lista maiores_vendas.
for i in range(len(vendas_janeiro)):
    if vendas_janeiro[i] == max(vendas_janeiro):
        maiores_vendas.append(i)
    else:
        pass
for i in range(len(vendas_fevereiro)):
    if vendas_fevereiro[i] == max(vendas_fevereiro):
        maiores_vendas.append(i)
    else:
        pass
for i in range(len(vendas_março)):
    if vendas_março[i] == max(vendas_março):
        maiores_vendas.append(i)
    else:
        pass
for i in range(len(vendas_abril)):
    if vendas_abril[i] == max(vendas_abril):
        maiores_vendas.append(i)
    else:
        pass
# Criando 4 novas variáveis que separam os elementos da lista maiores_vendas em 4 itens, para individualmente servir de índice para a próxima etapa.        
maior_venda_janeiro = maiores_vendas[0]                 
maior_venda_fevereiro = maiores_vendas[1]
maior_venda_março = maiores_vendas[2] 
maior_venda_abril = maiores_vendas[3]
# Finalmente, imprimindo o produto correspondente ao índice armazenado nas novas variáveis criadas.
print('O produto mais vendido de janeiro foi o produto', produtos[maior_venda_janeiro])
print('O produto mais vendido de fevereiro foi o produto', produtos[maior_venda_fevereiro])
print('O produto mais vendido de março foi o produto', produtos[maior_venda_março])
print('O produto mais vendido de abril foi o produto', produtos[maior_venda_abril])
# Esse exercício em especial foi muito difícil de resolver, não consegui achar uma forma ótima de resolvê-lo.       
# #c)
# Criando uma lista vazia onde serão armazenados todos os valores do crescimento percentual dos produtos no intervalo de tempo especificado.
crescimento_percentual = []
# Criando loop para adicionar os crescimentos percentuais entre janeiro e fevereiro na lista vazia.
for i in range(len(produtos)):
    crescimento_percentual.append(vendas_fevereiro[i]/vendas_janeiro[i])
# Criando variável que corresponde ao maior valor na lista entre os crescimentos percentuais AKA maior crescimento percentual.
maior_crescimento = max(crescimento_percentual)
# Criando um loop que percorre a lista de crescimentos percentuais, até achar o index do maior.
# Em seguida, ao obter o index do maior crescimento, nomeamos produto_c como a variável que tem o index desejado.
for i in range(len(crescimento_percentual)):
    if maior_crescimento == vendas_fevereiro[i]/vendas_janeiro[i]:
        produto_c = i
    else:
        pass
# Finalmente, nomeamos a variável resultado_c como sendo o produto que partilha o index do maior crescimento percentual através da variável produto_c.
resultado_c = produtos[produto_c]
# Imprimindo o resultado.
print('O produto com o maior crescimento percentual é o produto', resultado_c, 'com um crescimento percentual de', maior_crescimento*100,'%')
#d)
# Criando uma lista vazia onde serão armazenados os índices dos dos valores que apresentaram queda entre março e abril.
queda_vendas = []
# Criando um loop que percorre as listas de vendas de março e abril e, caso haja uma queda, salve o índice na lista vazia.
for i in range(len(produtos)):
    if vendas_março[i] > vendas_abril[i]:
        queda_vendas.append(i)
    else:
        pass
# Criando um loop que percorre a lista de produtos e, caso o produto compartilhe o mesmo índice da lista queda_vendas, imprima o resultado desejado.        
for i in range(len(produtos)):
    if i in queda_vendas:
        print('O produto',produtos[i],'teve uma queda nas vendas de março para abril')
    else:
        pass    
        
```
